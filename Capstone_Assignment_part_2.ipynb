```python
import pandas as pd
```

```python
#The data in the Table is imported and put in a data frame
ontario = pd.read_csv("Geospatial_Coordinates.csv")
ontario.head(11)
```

```python
#The previous data in part1 is imported and put in a data frame
toronto2 = pd.read_csv("toronto3.csv")
toronto2.head(11)
```

```python
#The 2 data frames are merged together, with one of the postal code column dropped
ontario_merge = pd.merge(toronto2, ontario, left_on='Postcode', right_on='Postal Code', how='left').drop('Postal Code', axis=1)
ontario_merge.rename(columns={'Postcode':'Postalcode'}, inplace=True)
ontario_merge.head(12)
```

```python
#The data is exported for use in the exploration and clustering section
ontario_merge.to_csv (r'C:\Users\OLUMEGBON ISMAIL\Desktop\Jupyter\ontario_merge.csv', index = None, header=True)
```
